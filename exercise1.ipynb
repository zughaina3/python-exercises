```python
def manipulate_string(s):
    reversed_str = s[::-1]
    every_second = s[::2]
    return reversed_str, every_second

def squares_of_evens(numbers):
    return [x**2 for x in numbers if x % 2 == 0]

def char_frequency(s):
    freq = {}
    for char in s:
        freq[char] = freq.get(char, 0) + 1
    return freq

def odd_indices(lst):
    return [idx for idx, val in enumerate(lst) if val % 2 != 0]

def rotate_list(lst, k):
    if not lst: return lst
    k = k % len(lst)
    return lst[-k:] + lst[:-k]

def lists_to_dict(keys, values):
    return dict(zip(keys, values))

def manual_iteration(lst):
    it = iter(lst)
    while True:
        try:
            print(next(it))
        except StopIteration:
            print("Reached end of list.")
            break

def flatten_list(nested):
    return [item for sublist in nested for item in sublist]

def is_palindrome(s):
    return s == s[::-1]

def second_largest(numbers):
    if len(numbers) < 2: return None
    first = second = float('-inf')
    for n in numbers:
        if n > first:
            second = first
            first = n
        elif n > second and n != first:
            second = n
    return second

print("Q1:", manipulate_string("AI Course"))
print("Q2:", squares_of_evens([1, 2, 3, 4, 5, 6]))
print("Q3:", char_frequency("programming"))
print("Q4:", odd_indices([10, 21, 32, 43, 54]))
print("Q5:", rotate_list([1, 2, 3, 4, 5], 2))
print("Q6:", lists_to_dict(['a', 'b', 'c'], [1, 2, 3]))
print("Q7:")
manual_iteration([10, 20, 30])
print("Q8:", flatten_list([[1, 2], [3, 4], [5]]))
print("Q9:", is_palindrome("radar"))
print("Q10:", second_largest([10, 20, 4, 45, 99]))
```
